```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

# Read the image (use a valid path to your image)
image = cv2.imread('Pictures/sunset.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()

# Reshape the image to a 2D array of pixels and 3 color values (RGB)
pixel_vals = image.reshape((-1, 3))

# Convert to float type
pixel_vals = np.float32(pixel_vals)

# Define criteria for k-means (stop either after 100 iterations or reaching 85% accuracy)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)

# Perform k-means clustering with the specified number of clusters (3)
# Random centers are chosen initially
k = 3
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Convert data back into 8-bit values
centers = np.uint8(centers)

# Map the centers to the labels
segmented_data = centers[labels.flatten()]

# Reshape data to the original image dimensions
segmented_image = segmented_data.reshape((image.shape))

# Display the segmented image
plt.imshow(segmented_image)
plt.show()
```
